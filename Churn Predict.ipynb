```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load and prepare the data again
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Handle categorical variables
binary_le = LabelEncoder()
df['gender'] = binary_le.fit_transform(df['gender'])
df['Partner'] = binary_le.fit_transform(df['Partner'])
df['Dependents'] = binary_le.fit_transform(df['Dependents'])
df['PhoneService'] = binary_le.fit_transform(df['PhoneService'])
df['PaperlessBilling'] = binary_le.fit_transform(df['PaperlessBilling'])
df['Churn'] = binary_le.fit_transform(df['Churn'])

# Create dummy variables
df = pd.get_dummies(df, columns=['InternetService', 'Contract', 'PaymentMethod',
                                'MultipleLines', 'OnlineSecurity', 'OnlineBackup',
                                'DeviceProtection', 'TechSupport', 'StreamingTV',
                                'StreamingMovies'])

# Risk scoring function
def calculate_churn_risk(row):
    risk_score = 0
    
    # Financial risk factors
    if pd.notnull(row['MonthlyCharges']) and row['MonthlyCharges'] > 74.44:
        risk_score += 0.3
    
    # Contract risk
    if 'Contract_Month-to-month' in df.columns and row['Contract_Month-to-month'] == 1:
        risk_score += 0.3
    
    # Tenure risk
    if pd.notnull(row['tenure']) and row['tenure'] < 12:
        risk_score += 0.2
    
    # Service risk
    if 'OnlineSecurity_No' in df.columns and row['OnlineSecurity_No'] == 1:
        risk_score += 0.2
    
    return risk_score

# Calculate risk scores
df['risk_score'] = df.apply(calculate_churn_risk, axis=1)

# Analyze risk distribution
print("\nRisk Distribution Analysis:")
print("--------------------------")
print(f"Total customers: {len(df)}")
print(f"High risk customers (score > 0.5): {len(df[df['risk_score'] > 0.5])}")
print(f"Critical risk customers (score > 0.7): {len(df[df['risk_score'] > 0.7])}")

# Calculate potential financial impact
avg_monthly_value = df['MonthlyCharges'].mean()
annual_value_per_customer = avg_monthly_value * 12
high_risk_revenue = df[df['risk_score'] > 0.5]['MonthlyCharges'].sum() * 12

print("\nFinancial Impact Analysis:")
print("-------------------------")
print(f"Average Annual Revenue per Customer: ${annual_value_per_customer:,.2f}")
print(f"Total Annual Revenue at Risk: ${high_risk_revenue:,.2f}")
print(f"Potential Savings (30% prevention): ${high_risk_revenue * 0.3:,.2f}")

# Create action priority segments
df['risk_category'] = pd.cut(df['risk_score'], 
                           bins=[0, 0.3, 0.5, 0.7, 1.0],
                           labels=['Low Risk', 'Medium Risk', 'High Risk', 'Critical Risk'])

print("\nAction Priority Segments:")
print("----------------------")
risk_distribution = df['risk_category'].value_counts()
for category, count in risk_distribution.items():
    print(f"{category}: {count} customers ({count/len(df)*100:.1f}%)")

# Generate recommendations for each risk category
print("\nTargeted Recommendations:")
print("-----------------------")
recommendations = {
    'Critical Risk': [
        "Immediate outreach for personalized retention offers",
        "Priority review of pricing and services",
        "Dedicated account manager assignment"
    ],
    'High Risk': [
        "Proactive contact for service review",
        "Targeted upgrade offers",
        "Contract term incentives"
    ],
    'Medium Risk': [
        "Regular satisfaction surveys",
        "Service usage analysis",
        "Promotional communications"
    ],
    'Low Risk': [
        "Standard engagement",
        "Periodic review",
        "Loyalty program enrollment"
    ]
}

for risk_level, actions in recommendations.items():
    print(f"\n{risk_level}:")
    for action in actions:
        print(f"- {action}")

# Create monitoring metrics
current_metrics = {
    'avg_monthly_charges': df['MonthlyCharges'].mean(),
    'month_to_month_ratio': df['Contract_Month-to-month'].mean() if 'Contract_Month-to-month' in df.columns else 0,
    'avg_risk_score': df['risk_score'].mean()
}

print("\nBaseline Monitoring Metrics:")
print("-------------------------")
for metric, value in current_metrics.items():
    print(f"{metric}: {value:.2f}")
```

```python
import matplotlib.pyplot as plt

# Create a comprehensive dashboard
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Risk Distribution Pie Chart
risk_sizes = [1559, 1528, 1032, 1556]  # Low, Medium, High, Critical
labels = ['Low Risk', 'Medium Risk', 'High Risk', 'Critical Risk']
colors = ['green', 'yellow', 'orange', 'red']
ax1.pie(risk_sizes, labels=labels, colors=colors, autopct='%1.1f%%')
ax1.set_title('Customer Risk Distribution')

# 2. Revenue at Risk Bar Chart
revenue_data = [739785.60, 2465952.00 - 739785.60]  # Potential savings vs remaining at-risk revenue
ax2.bar(['Potential Savings', 'Remaining At-Risk'], revenue_data, color=['green', 'red'])
ax2.set_title('Revenue Impact Analysis')
ax2.set_ylabel('Amount ($)')
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45)

# 3. Risk Score Distribution
ax3.hist(df['risk_score'], bins=20, color='blue', alpha=0.7)
ax3.set_title('Risk Score Distribution')
ax3.set_xlabel('Risk Score')
ax3.set_ylabel('Number of Customers')

# 4. Monitoring Metrics
metrics = ['Avg Monthly Charges', 'Month-to-Month Ratio', 'Avg Risk Score']
values = [64.76, 0.55, 0.46]
ax4.bar(metrics, values, color=['purple', 'blue', 'orange'])
ax4.set_title('Key Monitoring Metrics')
plt.setp(ax4.xaxis.get_majorticklabels(), rotation=45)

plt.tight_layout()
plt.show()

# Create action plan implementation framework
def create_intervention_plan():
    # Priority 1: Critical Risk Customers
    critical_risk_df = df[df['risk_category'] == 'Critical Risk'].copy()

    # Sort by monthly charges to prioritize high-value customers
    critical_risk_df = critical_risk_df.sort_values('MonthlyCharges', ascending=False)

    print("Immediate Action Plan (Top 10 Priority Customers):")
    print("------------------------------------------------")
    for idx, row in critical_risk_df.head(10).iterrows():
        print(f"Customer ID: {row['customerID']}")
        print(f"Monthly Charges: ${row['MonthlyCharges']:.2f}")
        print(f"Risk Score: {row['risk_score']:.2f}")
        print("Recommended Actions:")
        if row['MonthlyCharges'] > 74.44:
            print("- Review pricing structure")
        if row['Contract_Month-to-month'] == 1:
            print("- Offer contract term incentives")
        if row['tenure'] < 12:
            print("- Schedule immediate customer satisfaction review")
        print("------------------------------------------------")

# Create weekly monitoring template
def create_monitoring_template():
    monitoring_metrics = {
        'Daily Metrics': [
            'New high-risk customers identified',
            'Successful interventions',
            'Failed retention attempts',
            'Revenue at risk'
        ],
        'Weekly Metrics': [
            'Risk score distribution changes',
            'Average monthly charges trend',
            'Contract conversion rates',
            'Customer satisfaction scores'
        ],
        'Monthly Metrics': [
            'Model performance validation',
            'Risk factor importance analysis',
            'Revenue impact assessment',
            'Intervention strategy effectiveness'
        ]
    }

    print("\nMonitoring Dashboard Template:")
    print("-----------------------------")
    for period, metrics in monitoring_metrics.items():
        print(f"\n{period}:")
        for metric in metrics:
            print(f"- {metric}")

# Create automated alert system
def setup_alert_system():
    alert_thresholds = {
        'risk_score_threshold': 0.7,
        'monthly_charges_increase': 0.15,  # 15% increase
        'churn_probability_threshold': 0.8
    }

    # Example alert logic
    high_priority_alerts = df[
        (df['risk_score'] > alert_thresholds['risk_score_threshold']) &
        (df['MonthlyCharges'] > df['MonthlyCharges'].mean() * (1 + alert_thresholds['monthly_charges_increase']))
    ]

    print("\nHigh Priority Alerts:")
    print(f"Number of customers requiring immediate attention: {len(high_priority_alerts)}")

# Execute the action plans
create_intervention_plan()
create_monitoring_template()
setup_alert_system()
```
